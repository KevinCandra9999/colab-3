### Membuat barchart

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nama = ['jaja nana','jiji nini','juju nunu','jeje nene','jojo nono']
sales = [20, 30, 40, 50, 60]

datasales = pd.DataFrame({'nama' : nama, 'sales' : sales})
datasales
```

```python
plt.bar(x = datasales["nama"],
        height = datasales["sales"])
plt.show
```

### mengatur ukuran chart

```python
plt.figure(figsize=(9,6))
plt.bar(x = datasales["nama"],
        height = datasales["sales"])
plt.xticks(rotation=45)
plt.show
```

### menggunakan seaborn

```python
import seaborn as sns
sns.set()
```

### pengaturan title dan font

```python
plt.figure(figsize=(9,6))
plt.bar(x = datasales["nama"],
        height = datasales["sales"])
plt.title("Daftar Penjualan 2022", fontsize=16, fontweight="bold")
plt.ylabel("JUmlah Penjualan", fontsize=12)
plt.xlabel("Nama Sales", fontsize=12)
plt.xticks(rotation=45)
plt.show
```

### mewarnai chart

```python
plt.figure(figsize=(9,6))
plt.bar(x = datasales["nama"],
        height = datasales["sales"],
        color = ['red','yellow','blue','green','magenta'])
plt.title("Daftar Penjualan 2022", fontsize=16, fontweight="bold")
plt.ylabel("JUmlah Penjualan", fontsize=12)
plt.xlabel("Nama Sales", fontsize=12)
plt.xticks(rotation=45)
plt.show
```

### membuat piechart

```python
plt.figure(figsize=(9,6))
plt.pie(datasales["sales"], labels=datasales["nama"])
plt.show
```

### fungsi-fungsi untuk pie chart

```python
plt.figure(figsize=(9,6))
plt.pie(datasales["sales"], labels=datasales["nama"], shadow = True, colors=["blue","yellow","cyan","purple","lime"])
plt.title("Daftar Penjualan 2022", fontsize=16, fontweight="bold")
plt.legend(title="nama sales")
plt.show
```

### Line ploting

```python
data1 = [45,60,55,74,88]
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(data1)
plt.show()
```

### title, label x, label y

```python
fig=plt.figure(figsize=(10,5))
plt.title("Data Penjualan")
plt.xlabel("Bulan")
plt.ylabel("Penjualan dalam Jutaan")
plt.plot(data1)
plt.show()
```

### Subplot

```python
data2 = [50,60,55,90,76]
data3 = [60,73,45,66,79]
```

```python
fig=plt.figure(figsize=(10,5))
plt.title("Data Penjualan")
plt.xlabel("Bulan")
plt.ylabel("Penjualan (dalam Jutaan)")
chart = plt.subplot()
chart.plot(data1)
chart.plot(data2)
chart.plot(data3)
plt.show()
```

### argumen xlabel pada plot()

```python
bulan = ['Jan','Feb','Mar','Apr','Mei']
```

```python
fig=plt.figure(figsize=(10,5))
plt.title("Data Penjualan")
plt.xlabel("Bulan")
plt.ylabel("Penjualan (dalam Jutaan)")
chart = plt.subplot()
chart.plot(bulan, data1)
chart.plot(data2)
chart.plot(data3)
plt.show()
```

### line style, marker can color

```python
fig=plt.figure(figsize=(10,5))
plt.title("Data Penjualan")
plt.xlabel("Bulan")
plt.ylabel("Penjualan (dalam Jutaan)")
chart = plt.subplot()
chart.plot(bulan, data1, ls=':', marker='o', color='k')
chart.plot(data2, ls='-.', marker='^', color='r')
chart.plot(data3, ls='--', marker='*', color='m')
plt.show()
```

### multiple subplot

```python
fig=plt.figure(figsize=(10,15))

chart1  = plt.subplot(311)
chart2  = plt.subplot(312)
chart3  = plt.subplot(313)

chart1.plot(bulan, data1, ls=':', marker='o', color='k')
chart2.plot(data2, ls='-.', marker='^', color='r')
chart3.plot(data3, ls='--', marker='*', color='m')

plt.xlabel("Bulan")
plt.ylabel("Penjualan (dalam Jutaan)")

chart1.set(title = "Data Penjualan A")
chart2.set(title = "Data Penjualan B")
chart3.set(title = "Data Penjualan C")

plt.savefig('datachart')
plt.show()
```

### Pie Chart

```python
data = [25,35,15,5,20]
fig=plt.figure(figsize=(10,5))
plt.pie(data)
plt.title('Penjualan Sepeda')
```

### menambah label dan persen

```python
merek = ['A','B','C','D','E','F']
data = [25,35,15,5,20,10]
fig=plt.figure(figsize=(10,5))
plt.pie(data, labels=merek, autopct='%1.2f%%')
plt.title('Penjualan Sepeda')
```

### startangle dan explode

```python
merek = ['A','B','C','D','E','F']
data = [25,35,15,5,20,30]
explode=[0,0,0.5,0,0,0]
fig=plt.figure(figsize=(10,5))
plt.pie(data, labels=merek, autopct='%1.2f%%', explode=explode, startangle=90)
plt.title('Penjualan Sepeda')
```
