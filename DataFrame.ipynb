### MEngkombinasi kondisi

```python
pemainbola[(pemainbola.market_value.gt(30)) &
            (pemainbola.nationality == 'England')]
```

```python
pemainbola[(pemainbola.age.lt(30)) &
            (pemainbola.club == 'Arsenal') &
             (pemainbola.market_value.gt(30))]
```

### Memasukan kondisi kedalam variabel

```python
usia = pemainbola.age.lt(30)
klub = pemainbola.club == 'Arsenal'
market = pemainbola.market_value.gt(30)
pemainbola[(usia) &
          (klub) &
          (market)]
```

```python
usia = pemainbola.age.lt(30)
klub = pemainbola.club == 'Arsenal'
market = pemainbola.market_value.gt(30)
pemainbola[(usia) |
          (klub) |
          (market)]
```

### 2d indexing

```python
usia = pemainbola.age.lt(30)
klub = pemainbola.club == 'Arsenal'
market = pemainbola.market_value.gt(30)
pemainbola.loc[(usia) & (klub) & (market), ['name','age','club','market_value','nationality']]
```

### melakukan Sorting

```python
pemainbola.sort_values(by='age', ascending=False)
```

```python
usia = pemainbola.age.lt(30)
klub = pemainbola.club == 'Arsenal'
market = pemainbola.market_value.gt(30)
pemainbola.loc[(usia) & (klub) & (market), ['name','age','club','market_value','nationality']].sort_values(by='age', ascending=False)
```
